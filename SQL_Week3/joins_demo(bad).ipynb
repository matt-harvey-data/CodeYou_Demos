```python
import pandas as pd
```

```python
country_gdp = {'country': ['United States', 'Canada', 'Mexico', 'United Kingdom', 'France'],
              'capital': ['Washington DC', 'Ottawa', 'Mexico City', 'London', 'Paris'],
              'gdp': ['$27.36T', '$2.51T', '$1.41T', '2.54T Pounds', '$4.31T']}
```

```python
country_pop = {'head of state': ['Donald Trump', 'Justin Trudeau', 'Claudia Sheinbaum', 'Vladimir Putin', 'Xi Jinping', 'Narendra Modi'],
              'country': ['United States', 'Canada', 'Mexico', 'Russia', 'China', 'India'],
              'national population': ['346M', '40M', '132M', '146M', '1416M', '1503M']}
```

```python
capital_pop = {'capital': ['Washington DC', 'Ottawa', 'London', 'Moscow', 'Copenhagen', 'Oslo'],
              'capital_population': ['702k', '1451k', '9748k', '12712k', '1400k', '717k']}
```

```python
def df_builder(df):
    output = pd.DataFrame(df)
    return output
```

```python
country_gdp_df = df_builder(country_gdp)
```

```python
country_pop_df = df_builder(country_pop)
```

```python
capital_pop_df = df_builder(capital_pop)
```
